```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
# اسماء الاعمده
print(df.columns)
```

```python
print(df.shape)
```

```python
# الاعمدة التي تحوي في جميع قيمها nun نزيلها
df.dropna(how="all", axis = 1,inplace = True)
print(df.shape)
```

```python
print(df.columns)
```

```python
df.head()
```

```python
#بعض العمليات الااحصائية على القيم الرقمية في الداتا
df.describe()
```

```python
# نوع القيم الموجودة في كل عمود
df.dtypes
```

```python
# عدد قيم النان في كل سطر من كل عمود
print(df.isnull().sum())
```

```python
# نستبدل قيم النان بالمتوسط و عملنا هاد الشي من شان ما يكون الانزياح بالداتا بالمتوسط اقل ما يمكن
df["BasePay"].fillna(value=df["BasePay"].mean(), inplace=True)
df["OvertimePay"].fillna(value=df["OvertimePay"].mean(), inplace=True)
df["OtherPay"].fillna(value=df["OtherPay"].mean(), inplace=True)
df["Benefits"].fillna(value=df["Benefits"].mean(), inplace=True)
```

```python
# هون تخلصنا من قيم النان
print(df.isnull().sum())
```

```python
# القيام ببعض العمليات الاحصائية على رواتب الموظفين
print("Mean: ", df["BasePay"].mean())
print("Median: ", df["BasePay"].median())
print("Minimum: ", df["BasePay"].min())
print("Maximum: ", df["BasePay"].max())
print("Range: ", df["BasePay"].max() - df["BasePay"].min())
print("Standard Deviation: ", df["BasePay"].std())
print("Mode: ", df["BasePay"].mode())
```

```python
# عرض توزع الرواتب على عدد الموظفين
plt.hist(df["BasePay"])
plt.xlabel('Sallaries')
plt.ylabel('Employees')
plt.title('Distribution of salaries on every employee:', fontweight = "bold")
plt.show()
```
